```python
# Creating a histogram that depicts the distribution of Carbon Dioxide

plt.hist(
    df["co2"],
    bins=15,
    edgecolor="black",
    color="indianred",
)
plt.grid(True)
plt.xlabel("CO2 amount")
plt.ylabel("Frequency")
plt.title("Distribution of CO2")
plt.savefig("C02 Distrinution.png")
plt.show()
```

# Interpretation

The graph shows that CO2 amounts follow a normal distribution, with higher CO2 amounts within the 300-400 range. Fewer occurrences are observed as you move further away from this range. There is a slight skew to the right due to a few higher CO2 values beyond 500.

```python
# YOUR C
plt.hist(
    df["ghgScore"],
    bins=20,
    edgecolor="black",
    color="PaleVioletRed"
)

plt.grid(True)
plt.xlabel("ghgScore")
plt.ylabel("Frequency")
plt.title("ghgScore Distibution In Cars")
plt.savefig("ghgScore.png")
plt.show()
```

# Interpretation

The graph illustrates the distribution of greenhouse gas (ghgScores) among cars, highlighting that the majority fall within a mid-range GHG efficiency.
Most cars have a 4-7 score, which is relatively high considering the impact of the emitted gases 

```python
# YOUR CODE HERE
plt.hist(
    df["feScore"],
    edgecolor="black",
    color="DarkOrange"
)

plt.grid(True)
plt.xlabel("feScore")
plt.ylabel("Frequency")
plt.title("feScore Distribution")
plt.savefig("feScore.png")
plt.show()
```

# Interpretation

The graph shows the distribution of fuel efficiency scores (feScore) among cars, revealing that most vehicles fall within a moderate range of fuel efficiency. A significant number of cars have average scores, suggesting that fuel efficiency is a central consideration for many vehicles, though it has not reached exceptional levels for the majority.
